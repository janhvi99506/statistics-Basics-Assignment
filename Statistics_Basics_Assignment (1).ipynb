 Question 1: What is the difference between descriptive statistics and inferential statistics? Explain with examples.

 Ans :

**Descriptive Statistics** is used to **summarize and describe** the features of a dataset. It does not make any predictions or conclusions beyond the data collected.

**Examples of Descriptive Statistics:**
- Mean, Median, Mode
- Standard Deviation, Variance
- Frequency tables, Histograms, Bar charts

> Example: If we collect marks of 50 students and calculate the average mark → that is descriptive statistics.

---

**Inferential Statistics** is used to **make predictions or draw conclusions** about a population based on a sample of data. It goes beyond the immediate data.

**Examples of Inferential Statistics:**
- Hypothesis Testing
- Confidence Intervals
- Regression Analysis

> Example: If we survey 200 people and conclude that 60% of all Indians prefer tea over coffee → that is inferential statistics.

| Feature | Descriptive | Inferential |
|---------|------------|-------------|
| Purpose | Summarize data | Make predictions |
| Scope | Only collected data | Entire population |
| Tools | Mean, charts | Hypothesis tests |
| Example | Average salary of 100 employees | Predicting average salary of all employees |

```python
# Question 1 - Practical Example
import numpy as np

# Sample data: marks of 10 students
marks = [72, 85, 90, 60, 78, 88, 95, 70, 65, 80]

# --- Descriptive Statistics ---
print("===== DESCRIPTIVE STATISTICS =====")
print(f"Data        : {marks}")
print(f"Mean (Avg)  : {np.mean(marks):.2f}")
print(f"Median      : {np.median(marks)}")
print(f"Std Dev     : {np.std(marks):.2f}")
print(f"Min         : {np.min(marks)}")
print(f"Max         : {np.max(marks)}")

# --- Inferential Statistics (using scipy) ---
from scipy import stats

print("\n===== INFERENTIAL STATISTICS =====")
# 95% confidence interval for the population mean
confidence = 0.95
n = len(marks)
mean = np.mean(marks)
se = stats.sem(marks)
ci = stats.t.interval(confidence, df=n-1, loc=mean, scale=se)
print(f"Sample Mean           : {mean:.2f}")
print(f"95% Confidence Interval: ({ci[0]:.2f}, {ci[1]:.2f})")
print("Conclusion: We are 95% confident that the true population mean lies in this range.")
```

 Question 2: What is sampling in statistics? Explain the differences between random and stratified sampling.

Ans :

**Sampling** is the process of selecting a **subset (sample)** from a larger group (population) to study and draw conclusions about the entire population. It saves time, cost, and effort compared to studying every individual.

---

### Random Sampling:
Every member of the population has an **equal chance** of being selected. Selection is completely random.

> Example: Randomly picking 50 students from a school of 1000 students.

**Advantages:** Simple, unbiased  
**Disadvantages:** May not represent all subgroups

---

### Stratified Sampling:
The population is divided into **subgroups (strata)** based on a characteristic (like age, gender, class), and samples are taken from each stratum proportionally.

> Example: Dividing 1000 students into Class 9, 10, 11, 12 and then randomly picking from each class.

**Advantages:** Better representation of all groups  
**Disadvantages:** Requires prior knowledge of population groups

| Feature | Random Sampling | Stratified Sampling |
|---------|----------------|--------------------|
| Selection | Completely random | From each subgroup |
| Representation | May miss subgroups | All groups represented |
| Complexity | Simple | More complex |
| Best for | Homogeneous population | Heterogeneous population |

```python
# Question 2 - Practical Example
import random
import numpy as np

# Population: 100 students
population = list(range(1, 101))

# --- Random Sampling ---
random.seed(42)
random_sample = random.sample(population, 10)
print("===== RANDOM SAMPLING =====")
print(f"Population size : {len(population)}")
print(f"Random Sample (10): {sorted(random_sample)}")

# --- Stratified Sampling ---
print("\n===== STRATIFIED SAMPLING =====")
# Divide into 4 classes (strata)
class9  = list(range(1, 26))   # 25 students
class10 = list(range(26, 51))  # 25 students
class11 = list(range(51, 76))  # 25 students
class12 = list(range(76, 101)) # 25 students

# Pick proportional sample from each (2 from each = 8 total)
stratified_sample = (
    random.sample(class9, 2) +
    random.sample(class10, 2) +
    random.sample(class11, 2) +
    random.sample(class12, 2)
)

print(f"Strata         : Class 9, 10, 11, 12 (25 students each)")
print(f"Stratified Sample (8): {sorted(stratified_sample)}")
print("Each class is represented in the sample!")
```

## Question 3: Define mean, median, and mode. Explain why these measures of central tendency are important.

Ans :

**Measures of Central Tendency** describe the center or typical value of a dataset.

---

### 1. Mean (Average)
Sum of all values divided by the number of values.
- **Formula:** Mean = Sum of all values / Total number of values
- **Best for:** Normally distributed data without outliers
> Example: Marks [60, 70, 80] → Mean = (60+70+80)/3 = 70

### 2. Median (Middle Value)
The middle value when data is arranged in ascending order.
- If odd count → middle value
- If even count → average of two middle values
- **Best for:** Data with outliers (e.g., income data)
> Example: [10, 20, 30, 40, 100] → Median = 30

### 3. Mode (Most Frequent)
The value that appears most frequently in the dataset.
- A dataset can have one mode, multiple modes, or no mode
- **Best for:** Categorical data
> Example: [2, 3, 3, 5, 7] → Mode = 3

### Why are they important?
- They give a **quick summary** of a dataset
- Help in **comparing datasets**
- Used in **business decisions** (e.g., most sold product = mode)
- Foundation for **advanced statistical analysis**

```python
# Question 3 - Practical Example
import statistics
import numpy as np

data = [60, 70, 70, 80, 85, 90, 90, 90, 95, 100]

print("===== MEASURES OF CENTRAL TENDENCY =====")
print(f"Data   : {data}")
print()

# Mean
mean = np.mean(data)
print(f"Mean   : {mean}")
print(f"         (Sum={sum(data)}) / (Count={len(data)}) = {mean}")

# Median
median = np.median(data)
print(f"\nMedian : {median}")
print(f"         Sorted data middle value = {median}")

# Mode
mode = statistics.mode(data)
print(f"\nMode   : {mode}")
print(f"         {mode} appears most frequently in the data")

print("\n===== IMPORTANCE =====")
print("Mean   → Best for symmetric data (e.g., exam scores)")
print("Median → Best when outliers exist (e.g., income data)")
print("Mode   → Best for categorical data (e.g., popular product)")
```

Question 4: Explain skewness and kurtosis. What does a positive skew imply about the data?

Ans:

### Skewness:
Skewness measures the **asymmetry** of a distribution around its mean.

| Type | Skewness Value | Description |
|------|---------------|-------------|
| Symmetric (Normal) | = 0 | Mean = Median = Mode |
| Positive Skew | > 0 | Tail on right side |
| Negative Skew | < 0 | Tail on left side |

**Positive Skew (Right Skew):**
- The tail is on the **right side**
- **Mean > Median > Mode**
- Most values are **low**, but a few very high values pull the mean up
> Real Example: Income distribution — most people earn average income, but a few billionaires pull the average up.

**Negative Skew (Left Skew):**
- The tail is on the **left side**
- **Mean < Median < Mode**
> Example: Age at retirement — most people retire late, few retire early.

---

### Kurtosis:
Kurtosis measures the **sharpness of the peak** and **heaviness of tails** of a distribution.

| Type | Kurtosis | Description |
|------|----------|-------------|
| Mesokurtic | = 3 | Normal distribution |
| Leptokurtic | > 3 | Sharp peak, heavy tails |
| Platykurtic | < 3 | Flat peak, thin tails |

**High kurtosis** means more extreme outliers are present in the data.

```python
# Question 4 - Practical Example
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Positive skewed data (income-like)
np.random.seed(42)
pos_skew_data = np.random.exponential(scale=2, size=1000)

# Normal data
normal_data = np.random.normal(loc=5, scale=1, size=1000)

print("===== SKEWNESS & KURTOSIS =====")
print(f"\nPositive Skewed Data:")
print(f"  Skewness  : {stats.skew(pos_skew_data):.4f}  (> 0 means right skew)")
print(f"  Kurtosis  : {stats.kurtosis(pos_skew_data):.4f}")

print(f"\nNormal Data:")
print(f"  Skewness  : {stats.skew(normal_data):.4f}  (≈ 0 means symmetric)")
print(f"  Kurtosis  : {stats.kurtosis(normal_data):.4f}")

# Plot
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].hist(pos_skew_data, bins=40, color='salmon', edgecolor='black')
axes[0].set_title('Positive Skew (Right Skewed)\nTail on Right Side', fontsize=12)
axes[0].set_xlabel('Value')
axes[0].set_ylabel('Frequency')

axes[1].hist(normal_data, bins=40, color='skyblue', edgecolor='black')
axes[1].set_title('Normal Distribution\nSymmetric (Skewness ≈ 0)', fontsize=12)
axes[1].set_xlabel('Value')
axes[1].set_ylabel('Frequency')

plt.tight_layout()
plt.savefig('skewness_plot.png', dpi=100)
plt.show()
print("\nConclusion: Positive skew means most values are low, but few high values pull the mean up.")
```

## Question 5: Implement a Python program to compute the mean, median, and mode of a given list of numbers.

Ans:
- **Mean** = Sum of all values / Count of values
- **Median** = Middle value of sorted data
- **Mode** = Most frequently occurring value

We use Python's `statistics` module and `numpy` library to compute these values efficiently.

```python
# Question 5 - Practical
import statistics
import numpy as np
from collections import Counter

numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

print("===== MEAN, MEDIAN & MODE CALCULATOR =====")
print(f"Given Numbers : {numbers}")
print(f"Total Count   : {len(numbers)}")
print()

# Mean
mean = np.mean(numbers)
print(f"Mean   = Sum / Count = {sum(numbers)} / {len(numbers)} = {mean:.2f}")

# Median
sorted_nums = sorted(numbers)
median = np.median(numbers)
print(f"\nSorted : {sorted_nums}")
print(f"Median = Middle value = {median}")

# Mode
freq = Counter(numbers)
mode = statistics.multimode(numbers)
print(f"\nFrequency : {dict(sorted(freq.items()))}")
print(f"Mode      = Most frequent value(s) = {mode}")

print("\n===== FINAL RESULTS =====")
print(f"  Mean   : {mean:.2f}")
print(f"  Median : {median}")
print(f"  Mode   : {mode}")
```

## Question 6: Compute the covariance and correlation coefficient between the following two datasets.

Ans:

**Covariance** measures how two variables **change together**.
- Positive covariance → both increase together
- Negative covariance → one increases, other decreases
- **Limitation:** Value depends on the scale of data, hard to interpret magnitude

**Correlation Coefficient (Pearson's r)** is a **standardized** version of covariance.
- Always between **-1 and +1**
- **+1** → Perfect positive relationship
- **-1** → Perfect negative relationship
- **0** → No linear relationship

| Value | Meaning |
|-------|---------|
| 0.9 to 1.0 | Very strong positive |
| 0.5 to 0.9 | Strong positive |
| 0.0 to 0.5 | Weak positive |
| Negative | Inverse relationship |

```python
# Question 6 - Practical
import numpy as np
import matplotlib.pyplot as plt

list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

print("===== COVARIANCE & CORRELATION =====")
print(f"list_x = {list_x}")
print(f"list_y = {list_y}")
print()

# Covariance matrix
cov_matrix = np.cov(list_x, list_y)
covariance = cov_matrix[0][1]
print(f"Covariance Matrix:\n{cov_matrix}")
print(f"\nCovariance (x, y) = {covariance:.2f}")
print("Interpretation: Positive covariance → both variables increase together")

# Correlation coefficient
corr_matrix = np.corrcoef(list_x, list_y)
correlation = corr_matrix[0][1]
print(f"\nCorrelation Coefficient = {correlation:.4f}")
print("Interpretation: Very close to 1 → Very strong positive relationship")

# Scatter plot
plt.figure(figsize=(6, 4))
plt.scatter(list_x, list_y, color='blue', s=100, zorder=5)
plt.plot(list_x, list_y, color='red', linestyle='--', label=f'Correlation = {correlation:.4f}')
plt.title('Scatter Plot: list_x vs list_y')
plt.xlabel('list_x')
plt.ylabel('list_y')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig('correlation_plot.png', dpi=100)
plt.show()
```

## Question 7: Write a Python script to draw a boxplot and identify outliers.

Ans:

A **Boxplot** (Box-and-Whisker plot) visually shows the distribution of data using 5 key values:
1. **Minimum** – Smallest value (excluding outliers)
2. **Q1 (25th percentile)** – Lower quartile
3. **Median (Q2)** – Middle value
4. **Q3 (75th percentile)** – Upper quartile
5. **Maximum** – Largest value (excluding outliers)

**IQR (Interquartile Range)** = Q3 - Q1

**Outlier Detection Rule:**
- Lower Fence = Q1 - 1.5 × IQR
- Upper Fence = Q3 + 1.5 × IQR
- Any value **outside these fences** is an **outlier**

```python
# Question 7 - Practical
import numpy as np
import matplotlib.pyplot as plt

data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

print("===== BOXPLOT & OUTLIER DETECTION =====")
print(f"Data : {data}")
print()

# Calculate statistics
Q1  = np.percentile(data, 25)
Q2  = np.percentile(data, 50)
Q3  = np.percentile(data, 75)
IQR = Q3 - Q1

lower_fence = Q1 - 1.5 * IQR
upper_fence = Q3 + 1.5 * IQR

print(f"Q1 (25th percentile) = {Q1}")
print(f"Q2 / Median          = {Q2}")
print(f"Q3 (75th percentile) = {Q3}")
print(f"IQR = Q3 - Q1        = {IQR}")
print(f"\nLower Fence = Q1 - 1.5*IQR = {lower_fence}")
print(f"Upper Fence = Q3 + 1.5*IQR = {upper_fence}")

# Identify outliers
outliers = [x for x in data if x < lower_fence or x > upper_fence]
print(f"\nOutliers : {outliers}")
if outliers:
    print(f"  {outliers} is/are outside the whisker range → OUTLIER")
else:
    print("  No outliers found!")

# Draw Boxplot
plt.figure(figsize=(8, 5))
bp = plt.boxplot(data, vert=False, patch_artist=True,
                 boxprops=dict(facecolor='skyblue'),
                 medianprops=dict(color='red', linewidth=2),
                 flierprops=dict(marker='o', color='red', markersize=10))
plt.title('Boxplot with Outlier Detection', fontsize=13)
plt.xlabel('Values')
plt.grid(True, axis='x')
plt.tight_layout()
plt.savefig('boxplot.png', dpi=100)
plt.show()

print("\nResult: The red dot(s) in the boxplot represent outlier(s).")
```

## Question 8: E-commerce — Relationship between Advertising Spend and Daily Sales

Ans:

As a **data analyst in an e-commerce company**, I would use **Covariance and Correlation** to explore the relationship between advertising spend and daily sales:

**Step 1: Covariance**
- Calculate covariance to check if both variables move in the same direction.
- If positive → more ad spend = more sales (good sign!).

**Step 2: Correlation Coefficient**
- Since covariance is scale-dependent, I calculate Pearson's correlation.
- If r is close to +1 → strong positive relationship between ad spend and sales.

**Step 3: Decision**
- High positive correlation → increasing ad budget is likely to increase sales.
- Correlation does NOT mean causation — other factors may also influence sales.

**Business Insight:** If correlation is 0.99+, the company should confidently invest more in advertising.

```python
# Question 8 - Practical
import numpy as np
import matplotlib.pyplot as plt

advertising_spend = [200, 250, 300, 400, 500]
daily_sales       = [2200, 2450, 2750, 3200, 4000]

print("===== E-COMMERCE: AD SPEND vs DAILY SALES =====")
print(f"Advertising Spend : {advertising_spend}")
print(f"Daily Sales       : {daily_sales}")
print()

# Covariance
cov = np.cov(advertising_spend, daily_sales)[0][1]
print(f"Covariance               = {cov:.2f}")
print("  Positive → Both increase together (good signal!)")

# Correlation
corr = np.corrcoef(advertising_spend, daily_sales)[0][1]
print(f"\nCorrelation Coefficient  = {corr:.4f}")
print("  Very close to 1 → Very strong positive relationship")

# Business Insight
print("\n===== BUSINESS INSIGHT =====")
print("As advertising spend increases, daily sales increase proportionally.")
print("Recommendation: Increase advertising budget to boost daily sales.")

# Scatter plot
plt.figure(figsize=(7, 5))
plt.scatter(advertising_spend, daily_sales, color='green', s=120, zorder=5)
plt.plot(advertising_spend, daily_sales, color='orange', linestyle='--')
for i, (x, y) in enumerate(zip(advertising_spend, daily_sales)):
    plt.annotate(f'({x},{y})', (x, y), textcoords="offset points", xytext=(5, 5))
plt.title(f'Advertising Spend vs Daily Sales\nCorrelation = {corr:.4f}', fontsize=12)
plt.xlabel('Advertising Spend (₹)')
plt.ylabel('Daily Sales (₹)')
plt.grid(True)
plt.tight_layout()
plt.savefig('ad_sales_correlation.png', dpi=100)
plt.show()
```

## Question 9: Customer Satisfaction Survey — Distribution Analysis

Ans:

As a data analyst, before launching a new product, I would analyze customer satisfaction survey scores (scale 1-10) using:

**Summary Statistics I would use:**
| Statistic | Purpose |
|----------|---------|
| **Mean** | Average satisfaction score |
| **Median** | Middle score (resistant to outliers) |
| **Mode** | Most common score given |
| **Std Deviation** | How spread out the scores are |
| **Min / Max** | Range of scores |
| **Skewness** | Whether scores are left/right skewed |

**Visualizations I would use:**
- **Histogram** → Shows frequency distribution of scores
- **Boxplot** → Shows median, IQR, and outliers
- **Bar Chart** → Shows count of each score

**Business Decision:**
- Mean > 7 → Customers are generally satisfied → Good to launch
- High std dev → Mixed opinions → May need more market research

```python
# Question 9 - Practical
import numpy as np
import matplotlib.pyplot as plt
import statistics
from scipy import stats

survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

print("===== CUSTOMER SATISFACTION SURVEY ANALYSIS =====")
print(f"Survey Scores : {survey_scores}")
print(f"Total Responses: {len(survey_scores)}")
print()

# Summary statistics
mean   = np.mean(survey_scores)
median = np.median(survey_scores)
mode   = statistics.mode(survey_scores)
std    = np.std(survey_scores)
skew   = stats.skew(survey_scores)

print(f"Mean              : {mean:.2f}")
print(f"Median            : {median}")
print(f"Mode              : {mode}")
print(f"Std Deviation     : {std:.2f}")
print(f"Min / Max         : {min(survey_scores)} / {max(survey_scores)}")
print(f"Skewness          : {skew:.4f}")

# Histogram
fig, axes = plt.subplots(1, 2, figsize=(13, 5))

# Histogram
axes[0].hist(survey_scores, bins=range(1, 12), color='steelblue',
             edgecolor='black', align='left', rwidth=0.8)
axes[0].axvline(mean,   color='red',    linestyle='--', linewidth=2, label=f'Mean={mean:.2f}')
axes[0].axvline(median, color='green',  linestyle='-',  linewidth=2, label=f'Median={median}')
axes[0].set_title('Histogram of Customer Satisfaction Scores', fontsize=12)
axes[0].set_xlabel('Score (1-10)')
axes[0].set_ylabel('Frequency')
axes[0].set_xticks(range(1, 11))
axes[0].legend()
axes[0].grid(axis='y', alpha=0.5)

# Boxplot
axes[1].boxplot(survey_scores, patch_artist=True,
                boxprops=dict(facecolor='lightyellow'),
                medianprops=dict(color='red', linewidth=2))
axes[1].set_title('Boxplot of Survey Scores', fontsize=12)
axes[1].set_ylabel('Score')
axes[1].grid(axis='y', alpha=0.5)

plt.tight_layout()
plt.savefig('survey_histogram.png', dpi=100)
plt.show()

print("\n===== BUSINESS CONCLUSION =====")
print(f"Average score is {mean:.2f}/10 → Customers are {'satisfied' if mean >= 7 else 'not fully satisfied'}.")
print("Recommendation: Product launch looks promising based on survey data!")
```
